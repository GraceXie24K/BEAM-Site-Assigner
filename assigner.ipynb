```python
#Import modules
import numpy as np
import pandas as pd
import random
```

```python
def load_data(file_path):
    return pd.read_excel(file_path)#.dropna().reset_index(drop=True)

def get_sitetimes_list(decal):
    full_set = set()
    for lst in decal["Availabilities"]:
        for item in lst:
            full_set.add(item.strip())
    return full_set

def ohe_availabilities(decal, sitetimes):
    availbilities = decal['Availabilities']
    new_df = pd.DataFrame()
    for time in sitetimes:
        ohe = []
        for person in availbilities:
            ohe.append(time in person)
        new_df[time] = ohe
    full_df = decal.join(new_df)
    return full_df

def add_fake_drivers(full_df):
    lst = []
    for i in range(full_df.shape[0]):
        lst.append(random.random() <= 0.3)
    full_df["Driver"] = lst
    return full_df
    
```

```python
#Clean Decal Dataset
decal = load_data(r"data\DeCal Availability Confirmation (Responses).xlsx").rename(columns = {"To help us have the most updated information when making rearrangements, please check all times you are available to go to site. Remember to leave 30 minutes before and after for site transportation.":"Availabilities"})
decal['Availabilities'] = decal['Availabilities'].str.split(", ")
decal = add_fake_drivers(decal)
```

```python
decal[decal['Name'] == "Parker Lum"]
```

```python
#Get Sitetimes
sitetimes = get_sitetimes_list(decal)

#Clean SL Dataset
sl = load_data(r"data\SL Availability Sp 23 (Responses).xlsx").rename(columns = {"Please select all times you are available to lead a site.":"Availabilities", "Do you have access to a car that seats 5 people, including yourself, that you can drive to site? OR Are you willing and able to rent a Gig/Zipcar/etc. to drive to site?":"Driver"})

#Clean Staff Dataset
staff = load_data(r"data\Staff Availability Sp 23 (Responses).xlsx").rename(columns = {"Please select all times you are available to go to site.":"Availabilities","Are you willing and able to rent a Gig/Zipcar/etc. to drive to site?":"Driver"})
staff = staff[["Name","Email Address","Availabilities","Driver"]]
```

```python
full_df = ohe_availabilities(decal, sitetimes)
full_df = full_df.drop(columns = ["Timestamp","Can you make the site time you have been assigned?","Availabilities"])

full_df[full_df['Name'] == "Parker Lum"]
```

```python
sl_df = ohe_availabilities(sl, sitetimes)
sl_df = sl_df.drop(columns = ["Timestamp","Availabilities","What was the date of your most recent Tuberculosis test and clearance? (Leave blank if N/A)"])
sl_df.head()
```

```python
staff_df = ohe_availabilities(staff, sitetimes)
staff_df = staff_df.drop(columns = ["Availabilities"])
staff_df.head()
```

```python
staff_df.to_csv("Staff.csv")
sl_df.to_csv("SL.csv")
```

```python
staff_names = set(staff['Name'].tolist())
sl_names = set(sl['Name'].tolist())
full_staff_names = staff_names.union(sl_names)
full_staff_names
```

```python
criteria = []
for name in full_df['Name']:
    criteria.append(name not in full_staff_names)
decal_df = full_df[criteria]
decal_df
```
